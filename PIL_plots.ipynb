```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

sns.set_style('white')
```

```python
model_name = 'resnet18'
results = pd.read_csv('./results/birds200/pil/val/' + model_name + '_agg_fs')
sns.barplot(x='aug', y='top1', hue='agg', data=results)
if model_name == 'resnet18':
    plt.ylim(75, 80)
elif model_name == 'resnet50':
    plt.ylim(79, 82)

plt.title("Birds200, " + model_name)
```
