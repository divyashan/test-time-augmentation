# PIL

```python
SMALL_SIZE = 16
MEDIUM_SIZE = 20
BIGGER_SIZE = 24

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
#all pil results
datasets = ['flowers102', 'imnet']
model_names = ['inceptionv3','resnet18', 'MobileNetV2', 'resnet50']
# new_palette = sns.color_palette('tab10')[3:5] + sns.color_palette('tab10')[:3]
# sns.set_palette(new_palette)
sns.set_palette(sns.color_palette('tab10'))
plt.figure(figsize=(16, 16))
all_expanded_results = []
for j, dataset in enumerate(datasets):
    for model_idx in range(len(model_names)):
        plt.subplot(4, 4, j*4 + 1 + model_idx)
        model_name = model_names[model_idx]
        results = pd.read_csv("results/" + dataset + "/pil/val/" + model_name + "_agg_fs")
        all_expanded_results.append(results)
        orig_acc = np.mean(results[results['aug'] == 'orig'][y_col])
        other = results[results['aug'] == 'combo']
        ax = sns.barplot(x='aug', y=y_col, hue='agg', data=other, hue_order=hue_order)
        ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))

        ax.axhline(orig_acc, ls='--', color='black', linewidth=2, label='Raw')
        ax.get_legend().remove()
        if y_col == 'top1':
            if dataset == 'imnet':
                plt.ylim(69, 78)
            elif dataset == 'birds200':
                plt.ylim(75, 82)
            elif dataset == 'flowers102':
                plt.ylim(88.5, 94)
        else:
            if dataset == 'imnet':
                plt.ylim(88, 94)
            elif dataset == 'birds200':
                plt.ylim(93, 96)
            elif dataset == 'flowers102':
                plt.ylim(97, 99)
        plt.title(model_name_dict[model_name])
        if dataset == 'imnet' and model_idx == 3:
            plt.legend()
            L=plt.legend()
            #L.get_texts()[1].set_text('Max')
            L.get_texts()[1].set_text('Mean')
            L.get_texts()[2].set_text('GPS')
#             L.get_texts()[3].set_text('AugTTA')
#             L.get_texts()[4].set_text('ClassTTA')
#             L.get_texts()[5].set_text('Ours')
            L.get_texts()[3].set_text('Ours')

            #L.get_texts()[4].set_text('ClassTTA')

        if model_idx == 0:
            if y_col == 'top1':
                plt.ylabel(dataset_name_dict[dataset] + "\nTop-1 Classification Acc.")
            else:
                plt.ylabel(dataset_name_dict[dataset] + "Top-5 Classification Acc.")
        else:
            plt.yticks([])
            plt.ylabel("")
        plt.xlabel("")
        plt.xticks([0], "")
plt.tight_layout()
```

```python
datasets = ['flowers102', 'imnet']
model_names = ['inceptionv3','resnet18', 'MobileNetV2', 'resnet50']
plt.figure(figsize=(16, 16))
all_standard_results = []
# new_palette = sns.color_palette('tab10')[3:5] + sns.color_palette('tab10')[:3]
# sns.set_palette(new_palette)
for j, dataset in enumerate(datasets):
    for model_idx in range(len(model_names)):
        plt.subplot(4, 4, j*4 + 1 + model_idx)
        model_name = model_names[model_idx]
        try: 
            results = pd.read_csv("results/" + dataset + "/five_crop_hflip_scale/val/" + model_name + "_agg_fs")
        except:
            continue
        all_standard_results.append(results)
        orig_acc = np.mean(results[results['aug'] == 'orig'][y_col])
        other = results[results['aug'] == 'combo']
        if len(other) == 0:
            continue
        ax = sns.barplot(x='aug', y=y_col, hue='agg', data=other, hue_order=hue_order)
        ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))

        #ax = sns.barplot(x='aug', y=y_col, hue='agg', data=other)
        ax.axhline(orig_acc, ls='--', color='black', linewidth=2, label='Raw')
        ax.get_legend().remove()

        if y_col == 'top1':
            if dataset == 'imnet':
                plt.ylim(69, 78)
            elif dataset == 'birds200':
                plt.ylim(75.5, 82.5)
            elif dataset == 'flowers102':
                plt.ylim(88, 94)
        else:
            if dataset == 'imnet':
                plt.ylim(88, 94)
            elif dataset == 'birds200':
                plt.ylim(93, 96)
            elif dataset == 'flowers102':
                plt.ylim(97, 99)
                
        plt.title( model_name_dict[model_name])
        if dataset == 'imnet' and model_idx == 3:
            plt.legend()
            L=plt.legend()
            #L.get_texts()[1].set_text('Max')
            L.get_texts()[1].set_text('Mean')
            L.get_texts()[2].set_text('GPS')
#             L.get_texts()[3].set_text('AugTTA')
#             L.get_texts()[4].set_text('ClassTTA')
#             L.get_texts()[5].set_text('Ours')
            L.get_texts()[3].set_text('Ours')

        if model_idx == 0:
            if y_col == 'top1':
                plt.ylabel(dataset_name_dict[dataset] + "\nTop-1 Classification Acc.")
            else:
                plt.ylabel(dataset_name_dict[dataset] + "Top-5 Classification Acc.")
        else:
            plt.yticks([])
            plt.ylabel("")
        plt.xlabel("")
        plt.xticks([0], "")
plt.tight_layout()
```
