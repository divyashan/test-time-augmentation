```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

sns.set_style('white')
model_names = ['inceptionv3','resnet18', 'MobileNetV2', 'resnet50']
hue_order = ['mean', 'gps', 'ours']
model_name_dict = {'resnet18': 'ResNet-18', 'resnet50': 'ResNet-50', 
                   'MobileNetV2': 'MobileNetV2', 'inceptionv3': 'InceptionV3'}
dataset_name_dict = {'birds200': 'Birds200', 'flowers102':'Flowers102', 'imnet':'ImageNet'}
legend_name_dict = {'mean': 'Standard', 'gps': 'GPS', 'max': 'Max', 'ours': 'Ours'}
legend_labels = ['Raw', 'Standard', 'GPS', 'Ours']
y_col = 'top1'
```

```python
# imnet five_crop plots
for model_idx in range(len(model_names)):
    model_name = model_names[model_idx]
    results = pd.read_csv("results/imnet/five_crop_hflip_scale/val/" + model_name + "_agg_fs")
    orig_acc = np.mean(results[results['aug'] == 'orig'][y_col])
    other = results[results['aug'] == 'combo']
    plt.figure(figsize=(6, 4))
    ax = sns.barplot(x='aug', y=y_col, hue='agg', data=other, hue_order=hue_order)
    ax.axhline(orig_acc, ls='--', color='black', linewidth=2, label='Raw')
    if model_name == 'inceptionv3':
        if y_col == 'top1':
            plt.ylim(69, 73)
        else:
            plt.ylim(88, 91)
    elif model_name == 'resnet18':
        if y_col == 'top1':
            plt.ylim(68, 73)
        else:
            plt.ylim(88, 91)
    elif model_name == 'MobileNetV2':
        if y_col == 'top1':
            plt.ylim(77, 81)
        else:
            plt.ylim(92, 96)
    elif model_name == 'resnet50':
        if y_col == 'top1':
            plt.ylim(75, 79)
        else:
            plt.ylim(92, 94)
    plt.title("ImageNet, " + model_name_dict[model_name])
    if model_idx == 3:
        plt.legend()
    plt.ylabel("Top1 Classification Accuracy")
    plt.xlabel("")
    plt.xticks([0], "")
```

# PIL

```python
# Setting matplotlib settings

SMALL_SIZE = 17
MEDIUM_SIZE = 17
BIGGER_SIZE = 17

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
#all pil results
datasets = ['flowers102', 'birds200', 'imnet']
plt.figure(figsize=(16, 16))
all_expanded_results = []
for j, dataset in enumerate(datasets):
    for model_idx in range(len(model_names)):
        plt.subplot(4, 4, j*4 + 1 + model_idx)
        model_name = model_names[model_idx]
        results = pd.read_csv("results/" + dataset + "/pil/val/" + model_name + "_agg_fs")
        all_expanded_results.append(results)
        orig_acc = np.mean(results[results['aug'] == 'orig'][y_col])
        other = results[results['aug'] == 'combo']
        ax = sns.barplot(x='aug', y=y_col, hue='agg', data=other, hue_order=hue_order)
        ax.axhline(orig_acc, ls='--', color='black', linewidth=2, label='Raw')
        ax.get_legend().remove()
        if y_col == 'top1':
            if dataset == 'imnet':
                plt.ylim(69, 77.5)
            elif dataset == 'birds200':
                plt.ylim(75.5, 82)
            elif dataset == 'flowers102':
                plt.ylim(88.5, 94)
        else:
            if dataset == 'imnet':
                plt.ylim(88, 94)
            elif dataset == 'birds200':
                plt.ylim(93, 96)
            elif dataset == 'flowers102':
                plt.ylim(97, 99)
        plt.title( dataset_name_dict[dataset] + ", " + model_name_dict[model_name])
        if dataset == 'imnet' and model_idx == 3:
            plt.legend()
            L=plt.legend()
            #L.get_texts()[1].set_text('Max')
            L.get_texts()[1].set_text('Mean')
            L.get_texts()[2].set_text('GPS')
            L.get_texts()[3].set_text('Ours')
        if model_idx == 0:
            if y_col == 'top1':
                plt.ylabel("Top-1 Classification Acc.")
            else:
                plt.ylabel("Top-5 Classification Acc.")
        else:
            plt.yticks([])
            plt.ylabel("")
        plt.xlabel("")
        plt.xticks([0], "")
plt.tight_layout()
#plt.suptitle("Expanded Test-Time Augmentation Policy")
```

```python
#all pil results
datasets = ['flowers102', 'birds200', 'imnet']
plt.figure(figsize=(16, 16))
all_standard_results = []
for j, dataset in enumerate(datasets):
    for model_idx in range(len(model_names)):
        plt.subplot(4, 4, j*4 + 1 + model_idx)
        model_name = model_names[model_idx]
        results = pd.read_csv("results/" + dataset + "/five_crop_hflip_scale/val/" + model_name + "_agg_fs")
        all_standard_results.append(results)
        orig_acc = np.mean(results[results['aug'] == 'orig'][y_col])
        other = results[results['aug'] == 'combo']
        ax = sns.barplot(x='aug', y=y_col, hue='agg', data=other, hue_order=hue_order)
        ax.axhline(orig_acc, ls='--', color='black', linewidth=2, label='Raw')
        ax.get_legend().remove()

        if y_col == 'top1':
            if dataset == 'imnet':
                plt.ylim(69, 77.5)
            elif dataset == 'birds200':
                plt.ylim(75.5, 82)
            elif dataset == 'flowers102':
                plt.ylim(88.5, 94)
        else:
            if dataset == 'imnet':
                plt.ylim(88, 94)
            elif dataset == 'birds200':
                plt.ylim(93, 96)
            elif dataset == 'flowers102':
                plt.ylim(97, 99)
                
        plt.title( dataset_name_dict[dataset] + ", " + model_name_dict[model_name])
        if dataset == 'imnet' and model_idx == 3:
            plt.legend()
            L=plt.legend()
            #L.get_texts()[1].set_text('Max')
            L.get_texts()[1].set_text('Mean')
            L.get_texts()[2].set_text('GPS')
            L.get_texts()[3].set_text('Ours')
        if model_idx == 0:
            if y_col == 'top1':
                plt.ylabel("Top-1 Classification Acc.")
            else:
                plt.ylabel("Top-5 Classification Acc.")
        else:
            plt.yticks([])
            plt.ylabel("")
        plt.xlabel("")
        plt.xticks([0], "")
plt.tight_layout()
```

```python
expanded = pd.concat(all_expanded_results)
standard = pd.concat(all_standard_results)
```

```python
ours_expanded = expanded[expanded['agg'] == 'ours']['top1']
mean_expanded = expanded[(expanded['agg'] == 'mean') &(expanded['aug'] != 'orig')]['top1']
gps_expanded = expanded[(expanded['agg'] == 'gps')]['top1']
orig_expanded = expanded[(expanded['aug'] == 'orig')]['top1']

ours_standard = standard[standard['agg'] == 'ours']['top1']
mean_standard = standard[(standard['agg'] == 'mean') &(standard['aug'] != 'orig')]['top1']
gps_standard = standard[(standard['agg'] == 'gps')]['top1']
orig_standard = standard[(standard['aug'] == 'orig')]['top1']

```

```python
from scipy.stats import wilcoxon
print(wilcoxon(ours_expanded, gps_expanded))
print(wilcoxon(ours_expanded, mean_expanded))
print(wilcoxon(ours_expanded, orig_expanded))
```

```python
from scipy.stats import wilcoxon
print(wilcoxon(ours_standard, gps_standard))
print(wilcoxon(ours_standard, mean_standard))
print(wilcoxon(ours_standard, orig_standard))
```
